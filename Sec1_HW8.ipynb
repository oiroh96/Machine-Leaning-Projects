# 0.) Import and Clean data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
#drive.mount('/content/gdrive/', force_remount = True)
```

```python
df = pd.read_csv("bank-additional-full.csv", sep=';')
```

```python
df.head()
```

```python
df = df.drop(["default", "pdays",	"previous",	"poutcome",	"emp.var.rate",	"cons.price.idx",	"cons.conf.idx",	"euribor3m",	"nr.employed"], axis = 1)
df = pd.get_dummies(df, columns = ["loan", "job","marital","housing","contact","day_of_week", "campaign", "month", "education"],drop_first = True)
```

```python
df.head()
```

```python
y = pd.get_dummies(df["y"], drop_first = True)
X = df.drop(["y"], axis = 1)
```

```python
obs = len(y)
plt.bar(["No","Yes"],[len(y[y.yes==0])/obs,len(y[y.yes==1])/obs])
plt.ylabel("Percentage of Data")
plt.show()
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler().fit(X_train)

X_scaled = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#1.) Based on the visualization above, use your expert opinion to transform the data based on what we learned this quarter

```python
###############
###TRANSFORM###
###############
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler()

X_scaled,y_train=ros.fit_resample(X_scaled,y_train)
```

# 2.) Build and visualize a decision tree of Max Depth 3. Show the confusion matrix.

```python
dtree_main = DecisionTreeClassifier(max_depth = 3)
dtree_main.fit(X_scaled, y_train)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
plot_tree(dtree_main, filled = True, feature_names = X.columns, class_names=["No","Yes"])


#fig.savefig('imagename.png')
```

# 1b.) Confusion matrix on out of sample data. Visualize and store as variable

```python
y_pred = dtree_main.predict(X_test)
y_true = y_test
cm_raw = confusion_matrix(y_true, y_pred)
```

```python
class_labels = ['Negative', 'Positive']

# Plot the confusion matrix as a heatmap
sns.heatmap(cm_raw, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# 3.) Use bagging on your descision tree

```python
# placeholder for optimizing max depth
dtree=DecisionTreeClassifier(max_depth=3)
```

```python
bagging=BaggingClassifier(estimator=dtree,
                   n_estimators=100, 
                   max_samples=.5, 
                   max_features =1.)
```

```python
bagging.fit(X_scaled, y_train)

y_pred=bagging.predict(X_test)
```

```python
y_true = y_test
cm_raw = confusion_matrix(y_true, y_pred)
```

```python
class_labels = ['Negative', 'Positive']

# Plot the confusion matrix as a heatmap
sns.heatmap(cm_raw, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# 4.) Boost your tree

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
# placeholder for optimizing max depth
dtree=DecisionTreeClassifier(max_depth=3)
```

```python
boost=AdaBoostClassifier(estimator=dtree,
                   n_estimators=100, 
                    learning_rate=.1)


# max_features =1.
```

```python
boost.fit(X_scaled, y_train)

y_pred=boost.predict(X_test)
```

```python
y_true = y_test
cm_raw = confusion_matrix(y_true, y_pred)

class_labels = ['Negative', 'Positive']

# Plot the confusion matrix as a heatmap
sns.heatmap(cm_raw, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# 5.) Train a Logistic regression superlearner.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
dtree_main = DecisionTreeClassifier()
dtree_main.fit(X_scaled, y_train)
```

```python
base_predictions=[list(dtree_main.predict(X_scaled)), 
                  list(boost.predict(X_scaled)), 
                  list(bagging.predict(X_scaled))]
```

```python
base_predictions=np.array(base_predictions).transpose()
```

```python
super_learner=LogisticRegression()
```

```python
super_learner.fit(base_predictions,y_train)
```

```python
super_learner.coef_
```

### Interpretation

The coefficients [[14.39022048, 1.20781882, 0.37526745]] from a Super Learner model indicate the relative importance assigned to each base learner's predictions within the ensemble. Specifically:

1. The first base learner, with a coefficient of 14.39022048, is deemed highly influential, suggesting its predictions significantly impact the ensemble's output.
2. The second learner, assigned a coefficient of 1.20781882, also positively contributes to the ensemble's predictions but to a lesser extent than the first.
3. The third learner, with a coefficient of 0.37526745, has the smallest positive influence, indicating its predictions are the least weighted among the three.
